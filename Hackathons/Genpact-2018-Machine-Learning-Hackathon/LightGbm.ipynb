```python
for col in agg_cols:
    tmp = X_train.groupby([col]). \
        agg({'num_orders' : [np.mean, np.sum, np.min, np.max, np.std]}).reset_index()
    tmp.columns = [col, 'mean_customer_num_orders_'+col, 'total_customer_num_orders_'+col,'min_customer_num_orders_'+col, 'max_customer_num_orders_'+col, 
                  'std_customer_num_orders_'+col]
    X_train = pd.merge(X_train, tmp, how='left', on=[col]) 
    X_test = pd.merge(X_test, tmp, how='left', on=[col]) 
    test = pd.merge(test, tmp, how='left', on=[col]) 
```

```python
from sklearn.metrics import mean_squared_error, mean_squared_log_error
```

```python
def RMSLE(y, pred):
    return (mean_squared_log_error(y, pred)**0.5)*100
```

```python
def train_predict_reg(learner, X_train, y_train, X_test, y_test, test): 
    '''
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - sample_size: the size of samples (number) to be drawn from training set
       - X_train: features training set
       - y_train: income training set
       - X_test: features testing set
       - y_test: income testing set
    '''
    
    results = {}
    
    start = time() # Get start time
    learner.fit(X_train, y_train)
    end = time() # Get end time
    
    results['train_time'] = end - start
        
    start = time() # Get start time
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train)
    end = time() # Get end time
    
    results['pred_time'] = end - start
    results['RMSLE_train'] = RMSLE(y_train, np.where(predictions_train<0,0,predictions_train))
    results['RMSLE_test'] = RMSLE(y_test, np.where(predictions_test<0,0,predictions_test))
    results['preds_final'] = learner.predict(test)
       
    # Success
    print("{} trained".format(learner.__class__.__name__))
    print("Results: {}".format(results))
        
    # Return the results
    return results
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import make_scorer 
def kFoldCV(reg, X_train, y_train, X_test, y_test, parameters, cv_sets, test):
    grid_obj = GridSearchCV(reg, parameters, n_jobs=-1, cv=cv_sets, verbose=10, 
                            scoring = make_scorer(mean_squared_error, greater_is_better=False))
    grid_fit = grid_obj.fit(X_train, y_train)
    best_reg = grid_fit.best_estimator_
    best_score = grid_fit.best_score_
    print('best model: ', best_reg)
    best_predictions = best_reg.predict(X_test)
    
    results_cv = {}
   
    best_predictions_train = best_reg.predict(X_train)
    best_predictions_test = best_reg.predict(X_test)
    best_predictions_final = best_reg.predict(test)
   
    results_cv['preds_final'] = best_predictions_final
    results_cv['best_reg'] = grid_fit.best_estimator_
    results_cv['best_acc_train'] = RMSLE(y_train, np.where(best_predictions_train<0,0,best_predictions_train))
    results_cv['best_acc_test'] = RMSLE(y_test, np.where(best_predictions_test<0,0,best_predictions_test))
    
    return results_cv
```
